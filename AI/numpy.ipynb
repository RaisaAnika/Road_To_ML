Numpy arrays are faster and more convenient than lists
Numpy array allows to easily work with numerical data


So lets verify the claims ...

```python
#we use getsizeof function in the sys library to see the memory it is using
#so lets chek how much space a number takes
a = 10
print(sys.getsizeof(a)) #memory allocated to a
```

```python
##So lets check memory allocation differences

list1 = range(1000) ##creating a list of 1000 values
#considering it is a list of all integers we can get size of list1 as follows:
print(sys.getsizeof(a)*len(list1))
```

```python
a1 = np.arange(1000)
print(a1.size * a1.itemsize)
```

so see for same size numpy array is taking 8000bytes whereas list is taking 28000 bytes.

```python
##Now, lests validate that numpy arrays are fast
l1 = [1,20,3]
l2 = [10,2,30]
#lets do element wise sum
x = l1+l2 ##this concats the list does not add
print('concat list result',x)
##So we need loop to perform addition
result = [(x+y) for x,y in zip(l1,l2)]
print ('addition result:',result)
```

```python
#Now lets try with 1000000 element containing lists
size = 1000000

list2 = range(size)
list3 = range(size)

###Now lets first instantiate timer 
start = time.time()
##Now lets add the list2 and list3 element wise
result = [(x+y) for x,y in zip(list2,list3)]
##now, lets calculate time, we will deduct start timne from current time
time_taken = time.time()- start ##time is in microseconds
print (time_taken)
```

```python
##Now lets do similar for numpy
n1 = np.arange(size)
n2 = np.arange(size)

##now time calculation time
start2 = time.time()
result2 = n1 + n2
time__taken =time.time()-start2
print(time__taken)
```

SEE THERE IS A HUGE TIME DIFFERENCE FOR SIMILAR COMPUTATION AND WE CAN DERIVE THAT NUMPY IS FASTER.
ALSO, SEE UNLIKE IN LIST FOR DOING ELEMENT WISE ADDITION IN NUMPY WE DO NOT NEED LOOPS, BUT SIMPLE + OPERATOR DOES THE JOB!

SO WE CAN CONCLUDE NUMPY IS SUITABLE FOR BIG CALCULATIONS, THUS ITS POPULARITY FOR DATA SCIENCE.
